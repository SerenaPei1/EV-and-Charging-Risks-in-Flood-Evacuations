```python
import pandas as pd
import geopandas as gpd
import matplotlib.pyplot as plt
import numpy as np
import pathlib
```

#### Vehicle Registrations - Zip Codes

```python
### Chicago Zip Code Boundaries
filename = 'data/Boundaries - ZIP Codes/chicago_boundary_zip.shp'
df = gpd.read_file(filename)
df['zip'] = df['zip'].astype(np.int64)
df.set_index('zip')
```

```python
fig, ax = plt.subplots(figsize = (10,10))
plt.rcParams['figure.dpi'] = 100
df.plot(ax=ax, color='#FDFD96', edgecolor='black')  
```

```python
### Vehicle Registration - Zip Code
regisfile = 'EV Registration Data/chicago_vehicles_combined.csv'
reg_df= pd.read_csv(regisfile)

reg_df.columns = ['city','zip','Hybrid_Num', 'EV_Num']
merge_df = df.merge(reg_df, on='zip')
vehicle_gdf = gpd.GeoDataFrame(merge_df)
```

```python
### ACS - Zip Code
vehfile = 'EV Registration Data/vehicle_zip.csv'
veh_df= pd.read_csv(vehfile)
vehicle_gdf = vehicle_gdf.merge(veh_df, on='zip')
```

```python
vehicle_gdf
```

```python
vehicle_gdf.columns
```

```python
vehicle_gdf = vehicle_gdf[['GEO_ID','zip', 'EV_Num','Hybrid_Num', 'veh_cnt', 'geometry']]
```

```python
vehicle_gdf['Gasoline_Num'] = vehicle_gdf.veh_cnt - vehicle_gdf.EV_Num - vehicle_gdf.Hybrid_Num
vehicle_gdf['Pct_EV'] = vehicle_gdf.EV_Num/vehicle_gdf.veh_cnt*100
vehicle_gdf['Pct_Hybrid'] = vehicle_gdf.Hybrid_Num/vehicle_gdf.veh_cnt*100
vehicle_gdf['Pct_Gasoline'] = vehicle_gdf.Gasoline_Num/vehicle_gdf.veh_cnt*100
```

```python
vehicle_gdf = vehicle_gdf[['zip', 'EV_Num','Hybrid_Num', 'Gasoline_Num', 'veh_cnt', 'Pct_EV', 'Pct_Hybrid', 'Pct_Gasoline', 'geometry']]
vehicle_gdf.columns = ['ZIPCODE', 'EV_Num','Hybrid_Num', 'Gasoline_Num', 'TotVeh', 'Pct_EV', 'Pct_Hybrid', 'Pct_Gasoline', 'geometry']
```

```python
vehicle_gdf.to_file('chicago_vehicle data_zip.shp') 
vehicle_gdf.to_csv('chicago_vehicle data_zip.csv')
```

```python
fig, ax = plt.subplots(figsize = (10,10))
plt.rcParams['figure.dpi'] = 100
vehicle_gdf.plot(ax=ax, column = 'EV_Num', cmap='Greens', legend=True, edgecolor='black')
plt.title('Number of EV Registrations per December 2023')
```

```python
fig, ax = plt.subplots(figsize = (10,10))
plt.rcParams['figure.dpi'] = 100
vehicle_gdf.plot(ax=ax, column = 'Pct_Hybrid', cmap='Blues', legend=True, edgecolor='black')
```

```python
fig, ax = plt.subplots(figsize = (10,10))
plt.rcParams['figure.dpi'] = 100
vehicle_gdf.plot(ax=ax, column = 'Pct_Gasoline', cmap='Reds', legend=True, edgecolor='black')
```

### Vehicle Registrations - Census Tract

```python
### Chicago Census Tract Boundaries
filename = 'data/Boundaries - Census Tracts/chicago_boundary_tract.shp'
tract_df = gpd.read_file(filename)

fig, ax = plt.subplots(figsize = (10,10))
plt.rcParams['figure.dpi'] = 100
tract_df.plot(ax=ax, color='#FDFD96', edgecolor='black')  
```

```python
# Calculate Area for zip and tract (not actual area, but polygon area) 
vehicle_gdf['area_zip']=vehicle_gdf.area
tract_df['area_tract'] = tract_df.area
```

```python
data_merged = gpd.sjoin(tract_df, vehicle_gdf, how="left")
```

```python
data_merged
```

```python
data_merged = data_merged.drop_duplicates(subset=['geoid', 'name'], keep='last')
```

```python
fig, ax = plt.subplots(figsize = (10,10))
plt.rcParams['figure.dpi'] = 100
data_merged.plot(ax=ax, column = 'Pct_EV',cmap='Greens', legend=True, edgecolor='black')  
```

```python
fig, ax = plt.subplots(figsize = (10,10))
plt.rcParams['figure.dpi'] = 100
vehicle_gdf.plot(ax=ax, column = 'Pct_EV',cmap='Greens', legend=True, edgecolor='black')  
```

```python
### ACS - Census Tract

filename = 'data/Census Household Vehicle/acs2022_5yr_B08201_14000US17031283100.shp'
# download from: https://censusreporter.org/tables/B08201/
tract_data = gpd.read_file(filename)
```

```python
tract_data.head()
```

```python
tract_data.columns
```

```python
tract_data = tract_data[['geoid', 'name', 'B08201001', 'B08201002', 
       'B08201003', 'B08201004', 'B08201005','B08201006',
       'geometry']]
```

```python
tract_data.head()
```

```python
tract_data.columns = ['geoid','name','hh_total', 'hh_0veh','hh_1veh', 'hh_2veh', 'hh_3veh', 'hh_4veh','geometry']
```

```python
tract_data
```

```python
tract_data['veh_tot'] = (0*tract_data.hh_0veh + 1*tract_data.hh_1veh+ 2*tract_data.hh_2veh+ 3*tract_data.hh_3veh+ 4*tract_data.hh_4veh)
```

```python
tract_data = tract_data[:-2]
```

```python
fig, ax = plt.subplots(figsize = (10,10))
plt.rcParams['figure.dpi'] = 100
tract_data.plot(ax=ax, color='#FDFD96', edgecolor='black') 
```

```python
data_merged_census = data_merged[['geoid','Pct_EV', 'Pct_Hybrid', 'Pct_Gasoline','geometry']]
```

```python
data_merged_census.head()
```

```python
vehicle_census = tract_data.merge(data_merged_census, on='geoid')
```

```python
for i in range(len(vehicle_census)):
    try:
        if vehicle_census['veh_tot'][i] == 0:
            vehicle_census['Pct_EV'][i] = 0
            vehicle_census['Pct_Hybrid'][i] = 0
            vehicle_census['Pct_Gasoline'][i] = 0
    except:
        pass
```

```python
vehicle_census['Num_EV'] = vehicle_census.veh_tot*vehicle_census.Pct_EV/100
vehicle_census['Num_Hybrid'] = vehicle_census.veh_tot*vehicle_census.Pct_Hybrid/100
vehicle_census['Num_Gasoline'] = vehicle_census.veh_tot*vehicle_census.Pct_Gasoline/100
```

```python
vehicle_census.columns
```

```python
vehicle_census
```

```python
vehicle_census = vehicle_census[['geoid', 'name', 'hh_total', 'hh_0veh', 'hh_1veh', 'hh_2veh', 'hh_3veh',
       'hh_4veh', 'veh_tot', 'Pct_EV', 'Pct_Hybrid',
       'Pct_Gasoline', 'Num_EV', 'Num_Hybrid', 'Num_Gasoline','geometry_y']]
```

```python
vehicle_census.columns = ['geoid', 'name', 'hh_total', 'hh_0veh', 'hh_1veh', 'hh_2veh', 'hh_3veh',
       'hh_4veh', 'veh_tot', 'Pct_EV', 'Pct_Hybrid',
       'Pct_Gasoline', 'Num_EV', 'Num_Hybrid', 'Num_Gasoline','geometry']
```

```python
vehicle_census = gpd.GeoDataFrame(vehicle_census, geometry=vehicle_census.geometry, crs="EPSG:4326")
```

```python
fig, ax = plt.subplots(figsize = (10,10))
plt.rcParams['figure.dpi'] = 100
vehicle_census.plot(ax=ax, column = 'Num_EV',cmap='Greens', legend=True, edgecolor='black')  
```
